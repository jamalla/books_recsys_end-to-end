```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv(
    "BX-Books.csv",
    sep=";",
    encoding="latin-1",
    on_bad_lines="skip",   # instead of error_bad_lines=False
    low_memory=False       # optional, but often helpful for big CSVs
)
```

```python
books.head()
```

```python
books.describe()
```

```python
books.info()
```

```python
books.shape
```

```python
books = books[['ISBN','Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
```

```python
books.rename(columns={
    "Book-Title": "title",
    "Book-Author": "author",
    "Year-Of-Publication": "year_of_publication",
    "Publisher": "publisher",
    "Image-URL-S": "image_url_s",
    "Image-URL-M": "image_url_m",
    "Image-URL-L": "image_url_l"
}, inplace=True)
```

```python
books.head(3)
```

```python
users = pd.read_csv(
    "BX-Users.csv",
    sep=";",
    encoding="latin-1",
    on_bad_lines="skip",   # instead of error_bad_lines=False
    low_memory=False       # optional, but often helpful for big CSVs
)
```

```python
users.head()
```

```python
users.shape
```

```python
users.rename(columns={
    "User-ID": "user_id",
    "Location": "location",
    "Age": "age"
}, inplace=True)
```

```python
users.head(3)
```

```python
ratings = pd.read_csv(
    "BX-Book-Ratings.csv",
    sep=";",
    encoding="latin-1",
    on_bad_lines="skip",   # instead of error_bad_lines=False
    low_memory=False       # optional, but often helpful for big CSVs
)
```

```python
ratings.head()
```

```python
ratings = ratings.rename(columns={
    "User-ID": "user_id",
    "ISBN": "isbn",
    "Book-Rating": "book_rating"
})
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={
    "isbn": "ISBN",
    "book_rating": "rating"
}, inplace=True)
```

```python
ratings.head()
```

```python
print("Books shape:", books.shape)
print("Users shape:", users.shape)
print("Ratings shape:", ratings.shape)
```

```python
ratings['user_id'].value_counts()
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
number_of_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_of_ratings
```

```python
number_of_ratings = number_of_ratings.rename(columns={
    'rating': 'number_of_ratings'
})
```

```python
number_of_ratings
```

```python
final_ratings = ratings_with_books.merge(number_of_ratings, on='title')
```

```python
final_ratings.head(3)
```
